```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")

# Train the model
train_results = model.train(
    data="TumorDetection/data.yaml",  # path to dataset YAML
    epochs=20,  # number of training epochs
    imgsz=640,  # training image size
    device=0,  # device to run on, i.e. device=0 or device=0,1,2,3 or device=cpu
)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("runs/detect/train2/weights/best.pt")  # pretrained YOLO11n model

# Run batched inference on a list of images
results = model("TumorDetection/test/images/glioma_1025_jpg.rf.16f677de0fc84afa6b702f2fb8bdb3c2.jpg")  # return a list of Results objects

# Process results list
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    print(boxes)
```

```python
from ultralytics import YOLO
from ultralytics import SAM

# Load the YOLO model
yolo_model = YOLO("runs/detect/train2/weights/best.pt")  # pretrained YOLO model

# Run batched inference on a list of images
results = yolo_model("TumorDetection/test/images/meningioma_1152_jpg.rf.2c46dd587317f2b9c5e9a2c5de430451.jpg")  # return a list of Results objects

# Load the SAM model
sam_model = SAM("sam2_b.pt")

for result in results:
     class_ids = result.boxes.cls.int().tolist()  # noqa
     if len(class_ids):
         boxes = result.boxes.xyxy  # Boxes object for bbox outputs
         sam_results = sam_model(result.orig_img, bboxes=boxes, verbose=False, save=True, device=0)
```
